```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

# Create a synthetic regression dataset
X, y = make_regression(n_samples=10, n_features=1, noise=10, random_state=1)

# Create and fit a linear regression model
regression_model = LinearRegression()
regression_model.fit(X, y)

# Make predictions on the test data
y_pred = regression_model.predict(X)

```

```python
# Set figure size
plt.figure(figsize=(8, 5))

# Plot the data points, regression line, and projection
plt.scatter(X, y, color='blue', label='Target Values')


plt.xlabel('1D Input Values')
plt.ylabel('Continuous Target Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(8, 5))

# Plot the data points, regression line, and projection
plt.scatter(X, y, color='blue', label='Target Values')
plt.scatter(X, y_pred, color='red', label='Predicted Line')
plt.plot(X, y_pred, color='red', linewidth=1, label='Regression Line')

# Calculate and plot the projections
for x, y_actual, y_predicted in zip(X, y, y_pred):
    plt.plot([x, x], [y_actual, y_predicted], 'k--', linewidth=0.5)
    plt.text(x, (y_actual + y_predicted) / 2, f'{(y_actual - y_predicted):.2f}', color='purple')

plt.xlabel('1D Input Values')
plt.ylabel('Continuous Target Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
delta = y-y_pred

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(X, np.zeros(len(X)), color='red', linewidth=1, label='Regression Line')
plt.scatter(X, delta, color='orange', label='Delta for prediction')


# Calculate and plot the projections
for x, d in zip(X, delta):
    plt.plot([x, x], [0, d], 'k--', linewidth=0.5)
    plt.text(x, d, f'{(d):.2f}', color='purple')

plt.xlabel('1D Input Values')
plt.ylabel('Delta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
abs_delta = np.abs(y-y_pred)

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(X, np.zeros(len(X)), color='red', linewidth=1, label='Regression Line')
plt.scatter(X, abs_delta, color='orange', label='Absolute Delta for prediction')


# Calculate and plot the projections
for x, d in zip(X, abs_delta):
    plt.plot([x, x], [0, d], 'k--', linewidth=0.5)
    plt.text(x, d, f'{(d):.2f}', color='purple')

plt.xlabel('1D Input Values')
plt.ylabel('Absolute Delta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sum_abs = [np.sum(abs_delta[0:i+1]) for i in range(len(abs_delta))]

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(np.zeros(len(X)), sum_abs, 'k--', label='Sum of Absolute Delta')
plt.scatter(np.zeros(len(X)), sum_abs, color='orange', label='Delta for prediction')


# Calculate and plot the projections
for i in range(len(abs_delta)-1):
    plt.plot([0,0],[sum_abs[i],sum_abs[i+1]], 'k--', linewidth=0.5)
for i in range(len(abs_delta)):
    plt.text(0, sum_abs[i], f'{(abs_delta[i]):.2f}', color='purple')

# plt.xlabel('1D Input Values')
# plt.ylabel('Absolute Delta of Target Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
total_sum_abs = np.sum(abs_delta)

# Set figure size
plt.figure(figsize=(8, 5))

plt.plot([0,0], [0,total_sum_abs], 'k--', label='Sum of Absolute Delta')
plt.scatter([0,0], [0,total_sum_abs], color='black')
plt.text(0, total_sum_abs/2, f'{(total_sum_abs):.2f}', color='purple')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(8, 5))

plt.plot([0,0], [0,total_sum_abs/10], 'k--', label='Mean Absolute Error')
plt.scatter([0,0], [0,total_sum_abs/10], color='black')
plt.text(0, total_sum_abs/10/2, f'{(total_sum_abs/10):.2f}', color='purple')

plt.legend()
plt.grid(True)
plt.show()
```

# mse

```python
delta = y-y_pred

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(X, np.zeros(len(X)), color='red', linewidth=1, label='Regression Line')
plt.scatter(X, delta, color='orange', label='Delta for prediction')


# Calculate and plot the projections
for x, d in zip(X, delta):
    plt.plot([x, x], [0, d], 'k--', linewidth=0.5)
    plt.text(x, d, f'{(d):.2f}', color='purple')

plt.xlabel('1D Input Values')
plt.ylabel('Delta of Target Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sq_delta = (y-y_pred)**2

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(X, np.zeros(len(X)), color='red', linewidth=1, label='Regression Line')
plt.scatter(X, sq_delta, color='orange', label='Squared Delta for prediction')


# Calculate and plot the projections
for x, d in zip(X, sq_delta):
    plt.plot([x, x], [0, d], 'k--', linewidth=0.5)
    plt.text(x, d, f'{(d):.2f}', color='purple')

plt.xlabel('1D Input Values')
plt.ylabel('Squared Delta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sum_sq_delta = [np.sum(sq_delta[0:i+1]) for i in range(len(sq_delta))]

# Set figure size
plt.figure(figsize=(8, 5))
plt.plot(np.zeros(len(X)), sum_sq_delta, 'k--', label='Sum of Squared Delta')
plt.scatter(np.zeros(len(X)), sum_sq_delta, color='orange', label='Squared Delta for prediction')


# Calculate and plot the projections
for i in range(len(sum_sq_delta)-1):
    plt.plot([0,0],[sum_sq_delta[i],sum_sq_delta[i+1]], 'k--', linewidth=0.5)
for i in range(len(sum_sq_delta)):
    plt.text(0, sum_sq_delta[i], f'{(sq_delta[i]):.2f}', color='purple')

# plt.xlabel('1D Input Values')
# plt.ylabel('Absolute Delta of Target Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
total_sum_sq_delta = np.sum(sq_delta)

# Set figure size
plt.figure(figsize=(8, 5))

plt.plot([0,0], [0,total_sum_sq_delta], 'k--', label='Sum of Squared Delta')
plt.scatter([0,0], [0,total_sum_sq_delta], color='black')
plt.text(0, total_sum_sq_delta/2, f'{(total_sum_sq_delta):.2f}', color='purple')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(8, 5))

plt.plot([0,0], [0,total_sum_sq_delta/10], 'k--', label='Mean Square Error')
plt.scatter([0,0], [0,total_sum_sq_delta/10], color='black')
plt.text(0, total_sum_sq_delta/10/2, f'{(total_sum_sq_delta/10):.2f}', color='purple')

plt.legend()
plt.grid(True)
plt.show()
```

# RMSE

```python
rmse = np.sqrt(total_sum_sq_delta/10)
# Set figure size
plt.figure(figsize=(8, 5))

plt.plot([0,0], [0,rmse], 'k--', label='Root Mean Square Error')
plt.scatter([0,0], [0,rmse], color='black')
plt.text(0, rmse/2, f'{(rmse):.2f}', color='purple')

plt.legend()
plt.grid(True)
plt.show()
```
